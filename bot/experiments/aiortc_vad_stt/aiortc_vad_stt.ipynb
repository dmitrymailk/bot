```python
import soundfile as sf
from IPython.display import Audio, display
import av
import numpy as np

resampler = av.AudioResampler(
    format="s16",
    layout="mono",
    rate=16_000,
)
# data, sample_rate = sf.read("bot/experiments/aiortc_vad_stt/test.wav")
data, sample_rate = sf.read("bot/experiments/aiortc_vad_stt/slowdown_sample.wav")
# data = (data * 32_767).astype(np.int16)
# data[:,0]
print(data, sample_rate)
```

```python
import librosa



print(data[:, 0].shape)


data_16k = librosa.resample(
    y=data[:, 0] + data[:, 1],
    orig_sr=sample_rate,
    target_sr=16_000,
)
print(data_16k.shape)


display(Audio(data_16k, rate=16_000))
```

```python
import torchaudio
import torch

resampler = torchaudio.transforms.Resample(48_000, 16_000)
data_16k = (
    resampler(
        torch.tensor(
            data[:, 0],
            dtype=torch.float32,
        )
    )
    .squeeze()
    .numpy()
)
print(data_16k.shape)
display(Audio(data_16k, rate=16_000))
```
